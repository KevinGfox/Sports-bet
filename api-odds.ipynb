```python
import requests
import pandas as pd
```

```python
# Define access key, sport serie, region and odd market to the API request
APIkey = '6247ce2390bacb118999dd1a2254961f'
sport = 'soccer_italy_serie_a'
regions = 'eu'
markets = 'h2h'
```

```python
# Get request to the api
r = requests.get(f"https://api.the-odds-api.com/v4/sports/{sport}/odds/?apiKey={APIkey}&regions={regions}&markets={markets}")
```

```python
odds_draw=[]
for i in range(18):
    r = requests.get(f"https://api.the-odds-api.com/v4/sports/{sport}/odds/?apiKey={APIkey}&regions={regions}&markets={markets}")
    home = r.json()[i]['home_team']
    away = r.json()[i]['away_team']
    odds = r.json()[i]['bookmakers'][0]['markets'][0]['outcomes'][2]['price']
    date = r.json()[i]['commence_time']
    odds_draw.append([date,home,away,odds])
```

```python
df_odds_draw = pd.DataFrame(odds_draw, columns=['date','home_city','away_city','odds'])
```
